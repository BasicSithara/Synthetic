```python
from indicnlp.tokenize import indic_tokenize  
from indicnlp.normalize.indic_normalize import IndicNormalizerFactory
import warnings
```

```python
def e_tokenize(es):
  es=es.lower()
  es= es.translate(str.maketrans('', '', string.punctuation))
  e_token=[]
  #indic_string=clean_text(text)
 
  for t in indic_tokenize.trivial_tokenize(es): 
    #print(t)
    #e_token=en_tokens.split(',')
    e_token.append(t)
    #df_e_tokens.loc[len(df_e_tokens)] = e_token
  return e_token

def m_tokenize(ms):
  input_text = ms.translate(str.maketrans('', '', string.punctuation))
  factory=IndicNormalizerFactory()
  normalizer=factory.get_normalizer("ml",remove_nuktas=True)
  output_text=normalizer.normalize(input_text)
  m_token=[]
  for t in indic_tokenize.trivial_tokenize(output_text): 
    m_token.append(t)
  return(m_token)

def cmtoken_clean(words):
  step1 = words.replace("`", "")
  step2 = step1.replace('"', '')
  step3 = step2.replace('" ','')
  step4 = step3.replace(',  ',',')
  return step3.strip()

```

```python
refs = []
for i in range(len(df_reviews)):
  row=df_reviews.english_reviews[i]
  refs.append(e_tokenize(row))


mal = []
for i in range(len(df_reviews)):
  row=df_reviews.malayalam_reviews[i]
  mal.append(m_tokenize(row))

cm=[]
for i in range(len(df_reviews)):
  row=df_tokens.cm_tokens[i]
  cm_tokens=row
  #print("\n********CM Tokens****************\n",cm_tokens)
  cm_tokens =str(cm_tokens)[1:-1]
  #print("\n********CM Tokens1****************\n",cm_tokens)
  cm_tokens =cmtoken_clean(cm_tokens)
  cm.append(cm_tokens)
```

```python
from bert_score import score
```

```python
#################################### BERT Score between Mal and Eng reviews #########################################################
f1_scores=[]
p_scores=[]
r_scores=[]
for i in range(len(df_reviews)):
  print("\n***************************************** In Row ",i," ************************************************************\n")
  ts=0.6+(i*0.01)
  cands=mal[i]
  refs1=refs[i]
  p=len(cands)
  t=len(refs1)
  if p<t:
    cands.extend([' '] * (t-p))
  elif t<p:
    refs1.extend([' '] * (p-t))
  P, R, F1 = score(cands,refs1,model_type="bert-base-multilingual-cased")
  f1_mean='{}'.format(F1.mean())
  #print(f1_mean)
  f1_scores += [f1_mean]
  p_mean='{}'.format(P.mean())
  #print(Precision mean)
  p_scores += [p_mean]
  r_mean='{}'.format(R.mean())
  #print(REcall_mean)
  r_scores += [r_mean]
```

```python
f1=[eval(i) for i in  f1_scores]
p=[eval(i) for i in  p_scores]
r=[eval(i) for i in  r_scores]
print("\n",f1,"\n",p,"\n",r,"\n")
print(len(f1),"\n")
print(len(p),"\n")
print(len(r),"\n")
```

```python
import pandas as pd
import numpy as np
with open('/content/bert_score_eng_mal.csv', 'w') as file:
   pass 
print("Empty File Created Successfully")
a = np.array(f1)
b = np.array(p)
c=np.array(r)
df = pd.DataFrame({"f1" : a, "precision" : b,"recall":c})
df.to_csv("bert_score_eng_mal.csv", index=False)
```
