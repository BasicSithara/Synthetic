```python
from indicnlp.tokenize import indic_tokenize  
from indicnlp.normalize.indic_normalize import IndicNormalizerFactory
import warnings
```

```python
def e_tokenize(es):
  es=es.lower()
  es= es.translate(str.maketrans('', '', string.punctuation))
  e_token=[]
  #indic_string=clean_text(text)
 
  for t in indic_tokenize.trivial_tokenize(es): 
    #print(t)
    #e_token=en_tokens.split(',')
    e_token.append(t)
    #df_e_tokens.loc[len(df_e_tokens)] = e_token
  return e_token

def m_tokenize(ms):
  input_text = ms.translate(str.maketrans('', '', string.punctuation))
  factory=IndicNormalizerFactory()
  normalizer=factory.get_normalizer("ml",remove_nuktas=True)
  output_text=normalizer.normalize(input_text)
  m_token=[]
  for t in indic_tokenize.trivial_tokenize(output_text): 
    m_token.append(t)
  return(m_token)

def cmtoken_clean(words):
  step1 = words.replace("`", "")
  step2 = step1.replace('"', '')
  step3 = step2.replace('" ','')
  step4 = step3.replace(',  ',',')
  return step3.strip()

def t_tokenize(ts):
  ts=ts.lower()
  ts= ts.translate(str.maketrans('', '', string.punctuation))
  t_token=[]
  #indic_string=clean_text(text)
 
  for t in indic_tokenize.trivial_tokenize(ts): 
    #print(t)
    #e_token=en_tokens.split(',')
    t_token.append(t)
  return t_token
```

```python
refs = []
for i in range(len(df_reviews)):
  row=df_reviews.english_reviews[i]
  refs.append(e_tokenize(row))


mal = []
for i in range(len(df_reviews)):
  row=df_reviews.malayalam_reviews[i]
  mal.append(m_tokenize(row))

cm = []
for i in range(len(df_reviews.cm_reviews)):
  row=df_reviews.cm_reviews[i]
  cm.append(t_tokenize(row))

trans = []
for i in range(len(df_reviews.transliterated_cm_reviews)):
  row=df_reviews.transliterated_cm_reviews[i]
  trans.append(t_tokenize(row))
```

ROUGE stands for Recall-Oriented Understudy for Gisting Evaluation. It was introduced in this paper in 2004. This is a really popular metric that you'll definitely find in the literature around text summarization. The metric is based on calculating the syntactic overlap between candidate and reference summaries (or any other text pieces). Rouge-1, Rouge-2, Rouge-L, and Rouge-S are some commonly calculated numbers. Here, Rouge-1 calculates the overlap of unigram(individual word) between the candidate and reference text pieces. Rouge-2 calculates the overlap of bigrams(word pairs) between the candidate and reference text pieces. Rouge-L calculates the overlap of the longest co-occurring in sequence n-grams between candidate and reference text pieces. Rouge-S calculates the overlap of Skip-bigram(of any pair of words in their sentence order) between the candidate and reference text pieces.
