```python
def split_lines(t):
  return t.split('.')

def split_lines1(t):
  return t.split(' ')

import re
def untokenize(words):
    """
    Untokenizing a text undoes the tokenizing operation, restoring
    punctuation and spaces to the places that people expect them to be.
    Ideally, `untokenize(tokenize(text))` should be identical to `text`,
    except for line breaks.
    """
    text = ''.join(words)
    step1 = text.replace("`` ",'"').replace("''",'"').replace('. . .',  '...')
    step2 = step1.replace("(","(").replace(")", ") ")
    step3 = re.sub(r' ([.,:;?!%]+)([ \'"`])', r"\1\2", step2)
    step4 = re.sub(r' ([.,:;?!%]+)$', r"\1", step3)
    step5 = step4.replace("'","'").replace("n't", "n't").replace("can not", "cannot")
    step6 = step5.replace("`", "'")
    step7 = step6.replace(",", " ")
    step8 = step7.replace("'","")
    return step8.strip()

def cmtoken_clean(words):
  step1 = words.replace("`", "")
  step2 = step1.replace('"', '')
  step3 = step2.replace('" ','')
  step4 = step3.replace(',  ',',')
  return step3.strip()
'''
def cmlinesplit_clean(words):
  text = ' '.join(words)
  step1 = text.replace('  ', ' ')
  return step1.strip()
'''
```

```python
for r,c,m in zip(range(len(df_reviews.malayalam_reviews)),range(len(df_all_tokens.mal_tokens)),range(len(df_all_tokens.cm_tokens))):
  print("\n*****************************************************************************In row",r,"*****************************************************************************\n")
  en_rv=df_reviews.english_reviews[r]
  print("********English Review****************\n",en_rv)
  mal_rv=df_reviews.malayalam_reviews[r]
  print("********Malayalam Review****************\n",mal_rv)

  mal_rv_split=split_lines(mal_rv)

  mal_tokens=df_all_tokens.mal_tokens[c]
  mal_untokens=untokenize(mal_tokens)
  mal_rv_split1=split_lines1(mal_untokens)

  cm_tokens=df_all_tokens.cm_tokens[m]
  cm_tokens =str(cm_tokens)[1:-1]
  cm_tokens =cmtoken_clean(cm_tokens)

  cm_untokens=untokenize(cm_tokens)
  cm_linesplit1=split_lines1(cm_untokens)
  cm_linesplit=split_lines(cm_untokens)

  text = ' '.join(cm_linesplit1)
  step1 = text.replace('  ', ' ')
  cm_linesplit1=step1.strip()
  cm_linesplit1=split_lines1(cm_linesplit1)

  for i in range(len(mal_rv_split)):
    res=mal_rv_split[i].split(" ")
    splitsen_len=len(res)
    substr=res[splitsen_len-1].translate(str.maketrans('','', string.punctuation))
    inserttxt = '.'
    try:
     #display index
     idx=mal_rv_split1.index(substr)
     if substr in cm_linesplit1:
        id=cm_linesplit1.index(substr)
        cm_linesplit1[id] = cm_linesplit1[id] + inserttxt
    except ValueError:
             pass  # Could happen in face of bad user input
    if 'i' in cm_linesplit1:
        id=cm_linesplit1.index('i')
        cm_linesplit1[id] = 'ഞാൻ'
    try:
        if len(cm_linesplit1)>1 and '.' in cm_linesplit1[1]:
          cm_linesplit1[1]=cm_linesplit1[1].replace('.', '')
    except ValueError:
             pass  # Could happen in face of bad user input
  
  text = ' '.join(cm_linesplit1)
  #print(text)
  cm=untokenize(text)

  print("\n********Recreated Sentence****************\n",cm)
  df_reviews.cm_reviews[r]=cm

```
